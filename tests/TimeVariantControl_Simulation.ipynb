```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt

s = ct.tf('s')
```

# I Controller

```python
ki = 0.5
Ts = 2e-3

G = ki/s

constroller = ct.c2d(G, Ts, method="bilinear")

y = ct.step_response(constroller, 1)

plt.figure()
plt.plot(y[0], y[1])
plt.show()

print("const double expected[] = {", end="")
for i, yk in enumerate(y[1]):
    if (i == 0):
        print(f"{yk:.3f}", end="")
    else:
        print(f", {yk:.3f}", end="")
print("};", end="")
```

# D (Thamed) Controller

```python
kd = 6
omega = 10 * 2 * 3.1415
Ts = 2e-3

G = s * kd / (1 + s / omega)

constroller = ct.c2d(G, Ts, method="bilinear")

y = ct.step_response(constroller, 1)

plt.figure()
plt.plot(y[0], y[1])
plt.show()

print("const double expected[] = {", end="")
for i, yk in enumerate(y[1]):
    if (i == 0):
        print(f"{yk:.3f}", end="")
    else:
        print(f", {yk:.3f}", end="")
print("};", end="")
```

# PID Controller

```python
kp = 1
ki = 10
kd = 1
omega1 = 1 * 2 * 3.1415

Ts = 2e-3

G = (kp + ki / s + s * kd / (1 + s / omega1))

constroller = ct.c2d(G, Ts, method="bilinear")

y = ct.step_response(constroller, 1)

plt.figure()
plt.plot(y[0], y[1])
plt.show()

print("const double expected[] = {", end="")
for i, yk in enumerate(y[1]):
    if (i == 0):
        print(f"{yk:.3f}", end="")
    else:
        print(f", {yk:.3f}", end="")
print("};", end="")
```

# Lead-Lag Controller

```python
omega1 = 1 * 2 * 3.1415
omega2 = 10 * 2 * 3.1415

Ts = 2e-3

G = (1 + s / omega1) / (1 + s / omega2)

constroller = ct.c2d(G, Ts, method="bilinear")

y = ct.step_response(constroller, 1)

plt.figure()
plt.plot(y[0], y[1])
plt.show()

print("const double expected[] = {", end="")
for i, yk in enumerate(y[1]):
    if (i == 0):
        print(f"{yk:.3f}", end="")
    else:
        print(f", {yk:.3f}", end="")
print("};", end="")
```

# Notch

```python
omega = 1 * 2 * 3.1415
D = 0.2 # Dampening
W = 1 # width

Ts = 2e-3

G = (1 + 2 * W * D * s / omega + s**2 / omega**2) / (1 + 2 * W * s / omega + s**2 / omega**2)

_ = ct.bode(G, plot=True, dB=True)

constroller = ct.c2d(G, Ts, method="bilinear")

y = ct.step_response(constroller, 1)

plt.figure()
plt.plot(y[0], y[1])
plt.show()

print("const double expected[] = {", end="")
for i, yk in enumerate(y[1]):
    if (i == 0):
        print(f"{yk:.3f}", end="")
    else:
        print(f", {yk:.3f}", end="")
print("};", end="")
```
