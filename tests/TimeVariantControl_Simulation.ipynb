```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt

s = ct.tf('s')
```

# I Controller

```python
ki = 0.5
Ts = 2e-3

G = ki/s

constroller = ct.c2d(G, Ts, method="bilinear")

y = ct.step_response(constroller, 1)

plt.figure()
plt.plot(y[0], y[1])
plt.show()

print("const double expected[] = {", end="")
for i, yk in enumerate(y[1]):
    if (i == 0):
        print(f"{yk:.3f}", end="")
    else:
        print(f", {yk:.3f}", end="")
print("};", end="")
```

# D (Thamed) Controller

```python
kd = 6
omega = 10 * 2 * 3.1415
Ts = 2e-3

G = s * kd / (1 + s / omega)

constroller = ct.c2d(G, Ts, method="bilinear")

y = ct.step_response(constroller, 1)

plt.figure()
plt.plot(y[0], y[1])
plt.show()

print("const double expected[] = {", end="")
for i, yk in enumerate(y[1]):
    if (i == 0):
        print(f"{yk:.3f}", end="")
    else:
        print(f", {yk:.3f}", end="")
print("};", end="")
```

# PI Controller

```python
kp = 1.7 
ki = 3.1

Ts = 2e-3

G = (kp + ki / s)

constroller = ct.c2d(G, Ts, method="bilinear")

y = ct.step_response(constroller, 1)

plt.figure()
plt.plot(y[0], y[1])
plt.show()

print("const double expected[] = {", end="")
for i, yk in enumerate(y[1]):
    if (i == 0):
        print(f"{yk:.3f}", end="")
    else:
        print(f", {yk:.3f}", end="")
print("};", end="")
```
